```python
import pandas as pd
import seaborn as sns
```

## Input CSV and Run All

```python
csv = 'starter_title_train.csv'
```

```python
train_data = pd.read_csv(csv)
print(len(train_data))
train_data
```

```python
train_data = train_data.drop_duplicates()
train_data = pd.DataFrame(train_data)
print(len(train_data))
train_data
```

```python
count_0s = pd.DataFrame(train_data['views']==0)
count_0s = count_0s.value_counts()
under_100 = pd.DataFrame(train_data['views']<100)
under_100 = under_100["views"].value_counts()
under_1000 = pd.DataFrame(train_data['views']<1000)
under_1000 = under_1000["views"].value_counts()
under_10000 = pd.DataFrame(train_data['views']<10000)
under_10000 = under_10000["views"].value_counts()
under_100000 = pd.DataFrame(train_data['views']<100000)
under_100000 = under_100000["views"].value_counts()
under_500000 = pd.DataFrame(train_data['views']<500000)
under_500000 = under_500000["views"].value_counts()
under_1m = pd.DataFrame(train_data['views']<1000000)
under_1m = under_1m["views"].value_counts()
under_2_5m = pd.DataFrame(train_data['views']<2500000)
under_2_5m = under_2_5m["views"].value_counts()
under_10m = pd.DataFrame(train_data['views']<10000000)
under_10m = under_10m["views"].value_counts()
under_25m = pd.DataFrame(train_data['views']<25000000)
under_25m = under_25m["views"].value_counts()
under_50m = pd.DataFrame(train_data['views']<50000000)
under_50m = under_50m["views"].value_counts()
under_100m = pd.DataFrame(train_data['views']<100000000)
under_100m = under_100m["views"].value_counts()
under_500m = pd.DataFrame(train_data['views']<500000000)
under_500m = under_500m["views"].value_counts()
```

```python
less_100k = {'views':[]}
_100k_to_500k = {'views':[]}
_500k_to_1m = {'views':[]}
_1m_to_2_5m =  {'views':[]}
_2_5m_10m =  {'views':[]}
_10m_25m = {'views':[]}
_25m_50m =  {'views':[]}
_50m_100m = {'views':[]}
_100m_500m = {'views':[]}
above_500m = {'views':[]}

for i in train_data['views']:
    if i<100000:
        less_100k['views'].append(i)
        
    elif 100000<=i<500000:
        _100k_to_500k['views'].append(i)
        
    elif 500000<=i<1000000:
        _500k_to_1m['views'].append(i)
        
    elif 1000000<=i<2500000:
        _1m_to_2_5m['views'].append(i)
        
    elif 2500000<=i<10000000:
        _2_5m_10m['views'].append(i)
    
    elif 10000000<=i<25000000:
        _10m_25m['views'].append(i)
   
    elif 25000000<=i<50000000:
        _25m_50m['views'].append(i)
    
    elif 50000000<=i<100000000:
        _50m_100m['views'].append(i)
    
    elif 100000000<=i<500000000:
        _100m_500m['views'].append(i)
    
    else:
        above_500m['views'].append(i)
```

```python
less_100k_df = pd.DataFrame(less_100k)

_100k_to_500k_df = pd.DataFrame(_100k_to_500k)

_500k_to_1m_df = pd.DataFrame(_500k_to_1m)

_1m_to_2_5m_df = pd.DataFrame(_1m_to_2_5m)

_2_5m_10m_df = pd.DataFrame(_2_5m_10m)

_10m_25m_df = pd.DataFrame(_10m_25m)

_25m_50m_df = pd.DataFrame(_25m_50m)

_50m_100m_df = pd.DataFrame(_50m_100m)

_100m_500m_df = pd.DataFrame(_100m_500m)

above_500m_df = pd.DataFrame(above_500m)
```
