```python
# %%
import numpy as np
import pandas as pd
# %%
df = pd.read_csv("../raw_data/merge.csv")
```

```python
iqr = df[df['views'].between(df['views'].quantile(.2), df['views'].quantile(.8), inclusive=True)]
```

```python
# Computing IQR
Q1 = df['views'].quantile(0.25)
Q3 = df['views'].quantile(0.75)
IQR = Q3 - Q1
```

```python

# Filtering Values between Q1-1.5IQR and Q3+1.5IQR
filtered = df.query('(@Q1 - 1.5 * @IQR) <= df')
```
